## Exercício 6
Utilizando um loop, imprima a média da massa corporal de cada espécie de pinguins

````python
especies = penguins['species'].unique()

for especie in especies:
  filtrada = penguins.loc[penguins['species']== especie]
  print(especie,':', filtrada['body_mass_g'].mean())
````

## Exercício 7
Deseja-se catalogar os pinguins da base de dados e para isso precisamos criar um ID para cada um deles.
Uma das ideias é pegar o primeiro caractere de cada coluna da base, concatenar tudo e criar uma coluna de ID.
Escreva um código que faça isso.

````python
penguins.columns
````

````python
def primeiro_digito(input):
  primeiro = str(input)[0]
  return(primeiro)

aux = pd.DataFrame()

for coluna in penguins.columns:
  aux[coluna] = penguins[coluna].apply(primeiro_digito)

aux.head(3)
````

````python
penguins.head(3)
````

````python
aux['id'] = aux.apply(''.join, axis = 1)
penguins['id']= aux['id']
penguins
````
